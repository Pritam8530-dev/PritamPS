#### Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.

Ans:

```python
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.express as px
```

```python
df = sns.load_dataset('titanic')
```

```python
df
```

```python
from sklearn import preprocessing
```

```python
label_encoder = preprocessing.LabelEncoder()
```

```python
df['sex'] = label_encoder.fit_transform(df['sex'])
```

```python
px.scatter(df , x = df['age'] , y = df['fare'] , color = df['survived'])
```

```python
df
```

```python
df['sex'].unique()
```

```python
px.scatter(df , x = df['age'] , y = df['fare'] , color = df['survived'] , size = 10 * df['sex'])
```

#### Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

Ans:

```python
df_tips = sns.load_dataset('tips')
```

```python
df_tips
```

```python
fig = px.box(df_tips , x = 'time' , y = 'total_bill')
fig.show()
```

```python
fig1 = px.box(df_tips , x = 'time' , y = 'total_bill' , points = 'all')
fig1.show()
```

```python
fig2 = px.box(df_tips , x = 'time' , y = 'total_bill' , points = 'all' , color = 'smoker')
fig2.show()
```

#### Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.

Ans:

```python
px.histogram(df_tips , x = 'sex' , y = 'total_bill' , pattern_shape= 'smoker' , color = 'day')
```

#### Q5. What is Distplot? Using Plotly express, plot a distplot.

Ans:

Distplot :
    
    Distplot is a plot in plotly library.
    
    It is showing scatterplot as well as histogram.
    
    scatterplot showing scatterness of data in between range of dataset.
    
    histogram is showing on x axis is data dispersion and on y axis is its frequency.

```python
import plotly.figure_factory as ff
```

```python
import plotly.figure_factory as ff
import numpy as np


x1 = np.random.randn(200) - 2
x2 = np.random.randn(200)
x3 = np.random.randn(200) + 2
x4 = np.random.randn(200) + 4


hist_data = [x1, x2, x3, x4]

group_labels = ['Group 1', 'Group 2', 'Group 3', 'Group 4']

# Create distplot with custom bin_size
fig = ff.create_distplot(hist_data, group_labels, bin_size=.2)
fig.show()
```
