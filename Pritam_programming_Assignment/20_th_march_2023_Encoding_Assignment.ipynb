#### Q2. What is nominal encoding? Provide an example of how you would use it in a real-world scenario.

Ans:

Nominal Encoding or OneHotEncoding is the process where it convert / encode the categorical data into numerical data.

OneHotEncoding is used when categorical feature contains large number of categories.

typically categories is greater than equal to 5.

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
df['day'].value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
```

```python
sex_encode = encoder.fit_transform(df[['sex']]).toarray()
```

```python
import pandas as pd
df_sex_encode = pd.DataFrame(sex_encode , columns = encoder.get_feature_names_out())
```

```python
df_sex_encode
```

```python
time_encode = encoder.fit_transform(df[['time']]).toarray()
df_time_encode = pd.DataFrame(time_encode , columns = encoder.get_feature_names_out())
```

```python
df_time_encode
```

```python
day_encode = encoder.fit_transform(df[['day']]).toarray()
df_day_encode = pd.DataFrame(day_encode , columns = encoder.get_feature_names_out())
```

```python
df_day_encode
```

```python
smoker_encode = encoder.fit_transform(df[['smoker']]).toarray()
df_smoker_encode = pd.DataFrame(smoker_encode , columns = encoder.get_feature_names_out())
```

```python
df_encode = pd.concat([df , df_sex_encode ,df_smoker_encode , df_day_encode , df_time_encode ] , axis = 1)
```

```python
df_encode
```
