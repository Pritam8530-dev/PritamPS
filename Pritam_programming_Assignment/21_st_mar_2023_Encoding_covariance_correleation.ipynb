#### Q1. What is the difference between Ordinal Encoding and Label Encoding? Provide an example of when you might choose one over the other.

Ans:

LabelEncoding :
    
    1.Encode the categorical data into numerical data.
    
    2.It is used when the data is not in order.
    
    eg : low , medium , hot.
        
         male , female
        
        
    3.LabelEncoding is used when categorical feature contain less number of categories.
    
    4. it is not a natural ordering between the categories of a categorical variable.
    
    5.It is not create a dummy variables like onehotEncoding.
    
    6.It is convert categorical data into numerical data in that variable itself.
    
    
    
OrdinalEncoding :
    
    1.Encode the categorical features into numerical feature.
    
    2.It is used when data is in order.
    
    eg : Excellent , Good ,satisfactory
        
        
    3. It is a natural ordering between the categories of a categorical variable.
    
    4.OrdinalEncoding is used when categorical feature contain less categories as well that data into order.
    
    5.It is not create a dummy variables like OnhotEncoding.
    
    6.It convert categorical data into numerical data in that variable itself.

#### Q2. Explain how Target Guided Ordinal Encoding works and provide an example of when you might use it in a machine learning project.

Target Guided Ordinal Encoding :
    
    Target Guided Ordinal Encoding is the encoding technique.
    
    which is used when categorical variable has dependency / relation / strongly associated with target variable.

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

Here we have dependency of time with total_bill

```python
df.groupby(['time'])['total_bill'].mean()
```

```python
mean = df.groupby(['time'])['total_bill'].mean().to_dict()
```

```python
mean 
```

```python
df['enoded_time'] = df['time'].map(mean)
```

#### Q3. Define covariance and explain why it is important in statistical analysis. How is covariance calculated?

Ans:

covariance :
    
    covariance is the one variable changes with other variable.
    
    covariance is measures strength of relationship and direction of relationship and degree of assocoation.
    
    when x is increases then y is increses.
    
    when x is decreses then y is decreases.
    
    when x is increases then y is decreases.
    
    when x is decreases then y is increases.
    
    covariance is the spread of data.
    
    types of covariance :
        
        1.positive covariance :
            
            when x is increases then y is increases.
            
            when x is decreases then y is decreases.
            
            
        2.Negative covariance :
            
            when x is increases then y is decreases.
            
            when x is decreases then y is increases.
            
            
            
    cov(x , y) = (x - sample mean of x) * (y - sample mean of y) / (n - 1)
    
    
    x = Data points of x
    
    y = Data points of y
    
    n = number of samples
    
    
In python , numpy library contains cov function with the help of that we calculate covariance.

           or with the help of dataframe we can also calculate covariance.

eg:
