##### Use the following data:

Ans: 

The subplots() function in pyplot module of matplotlib library is used to create a figure and a set of subplots.

The subplot() function takes three arguments that describes the layout of the figure.

The layout is organized in rows and columns, which are represented by the first and second argument.

The third argument represents the index of the current plot.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
```

```python
plt.subplots(1 ,2)
plt.plot(x , y)
```

```python
plt.subplots(2,1)
plt.plot(x,y)
```

#### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
#### import numpy as np
#### company = np.array(["Apple", "Microsoft", "Google", "AMD"])
#### profit = np.array([3000, 8000, 1000, 10000])

Ans:

Barplot :
    
    
    Barplot is one of important plot in matplot library.
    
    It is used for data analysis and data visualisation.
    
    A bar plot or bar chart is a graph that represents the category of data with 
    
    rectangular bars with lengths and heights that is proportional to the values which they represent. 
    
    The bar plots can be plotted horizontally or vertically. 
    
    A bar chart describes the comparisons between the discrete categories. 
    
    One of the axis of the plot represents the specific categories being compared, 
    
    while the other axis represents the measured values corresponding to those categories.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
c = ['magenta','red','pink','orange']
plt.bar(company , profit , color = c)
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Company vs Profit')
plt.show()
```

```python
plt.figure(figsize = (10 , 10))
c = ['magenta','red','pink','orange']
plt.bar(company , profit , color = c ,align = 'edge')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Company vs Profit')
plt.legend()
```

```python
c = ['magenta','maroon','pink','orange']
plt.bar(company , profit , color = c ,align = 'center')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Company vs Profit')
plt.grid()
```

```python
plt.figure(figsize = (10 , 10))
c = ['magenta','maroon','pink','orange']
plt.barh(company , profit , color = c ,align = 'center')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Company vs Profit')
plt.show()
```

```python
plt.figure(figsize = (10 , 5))
c = ['magenta','maroon','pink','orange']
plt.barh(company , profit , color = c ,align = 'center')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Company vs Profit')
plt.grid()
```

#### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
#### box1 = np.random.normal(100, 10, 200)
#### box2 = np.random.normal(90, 20, 200)

Ans:

Boxplot : 
    
    Boxplot is a plot in matplotlib library.
    
    Boxplot is used for to detect outliers.
    
    Outliers are nothing but anamoly.
    
    Boxplot is used for Anamoly detection.
    
    outliers are those data points which are on abnormal distance.
    
    The data points which is apart of low limit and high limit.
    
    IQR [InterQuartile Range] = Q3 - Q1
    
    whisker width = 1.5
    
    low_limit = Q1 - (whisker width * IQR)
    
    High_limit = Q3 + (whisker width * IQR)

```python
import pandas as pd
```

```python
box1 = pd.DataFrame(np.random.normal(100, 10, 200))
box2 = pd.DataFrame(np.random.normal(90, 20, 200))
```

```python
box1.boxplot()
plt.title('Box plot of box1')
```

```python
box2.boxplot()
plt.title('Box plot of box2')
```

```python
plt.boxplot(box1)
plt.title('Box plot of box1')
```

```python
plt.boxplot(box2)
plt.title('Box plot of box2')
```
