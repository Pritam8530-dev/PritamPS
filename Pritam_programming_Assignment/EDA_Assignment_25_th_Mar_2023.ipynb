#### Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs

Ans:

```python
df = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')
```

```python
df.info()
```

```python
df['Installs'].unique()
```

```python
df['Installs'] = df['Installs'].str.split('+').str[0]
```

```python
df['Installs'] =  df['Installs'].str.replace(',' , '')
```

```python
df['Installs']
```

```python
df['Installs'].unique()
```

```python
df[df['Installs'].isnull()]
```

```python
df[df['Installs'] == 'Free']
```

```python
df['Installs'].mode()[0]
```

```python
df['Installs'].replace('Free' ,df['Installs'].mode()[0] , inplace = True )
#df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0 , '2 stops' : 2 , '1 stop': 1 , '3 stops':3 , '4 stops': 4,np.nan : 1 })
```

```python
df[df['Installs'] =='Free']
```

```python
df.isnull().sum()
```

```python
df['Installs'] = df['Installs'].astype(int)
```

```python
df.columns
```

```python
df_new = df.groupby(['App'])['Installs'].sum().sort_values(ascending = False).reset_index()
df_new['Installs'] = df_new['Installs'] / 1000000000 
#df.groupby(['Category'])['Installs'].sum().sort_values(ascending = False).reset_index()
```

```python
df_new
```

```python
df1 = df_new[:10]
```

```python
df1
```

```python
plt.figure(figsize = (30 , 10))
sns.barplot(x = 'App' , y = 'Installs' , data = df1)
plt.ylabel('No.of Installations in billion')
plt.title('Most popular 10 apps')
```
