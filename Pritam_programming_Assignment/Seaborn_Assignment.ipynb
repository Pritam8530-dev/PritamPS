#### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Ans:

1.countplot :
    
    countplot is used for to show relation of x axis data points with respect to y axis.
    
    eg : On x axis species and on y axis count is there.
        
        so countplot showing count of species.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df
```

```python
sns.countplot(df['species'] , data = df)
```

2.pairplot : 
    
    pairplot is use for showing data distribution of each column of dataset with respect to other columns of dataset

```python
sns.pairplot(df , hue = 'species')
```

3.scatterplot :
    
    scatterplot is use for to show how data get scatter in between range of dataset in two diffrent attribute.
    
    from that we understand there is positive covariance or negative covariance in that two perticular attribute.

```python
sns.scatterplot(x = df['sepal_length'], y = df['petal_length'] , data = df , hue = 'species' , style = 'species')
```

From this scatterplot we understood there is positive covariance in versicolor and verginica.

4.relplot:
    
    replot is use for to show relation between two attributes .
    
    how data get scatter in between range of dataset in two diffrent attribute.

    from that we understand there is positive covariance or negative covariance in that two perticular attribute.

```python
sns.relplot(x = df['sepal_length'], y = df['petal_length'] , data = df , hue = 'species' , style = 'species')
```

5.jointplot:
    
    From jointplot we get scatterplot as well as we get histogram.
    
    from scatterplot we get data relation between two attributes .
    
    and from histogram we get relation of data dispersion and its frequency.

```python
sns.jointplot(x = df['sepal_width'] , y = df['petal_width'] , data = df)
```

Boxplot:
    
    Boxplot is a plot in matplotlib library.

    Boxplot is used for to detect outliers.

    Outliers are nothing but anamoly.

    Boxplot is used for Anamoly detection.

    Outliers are those data points which are on abnormal distance.

    The data points which is apart of low limit and high limit.

    IQR [InterQuartile Range] = Q3 - Q1

    whisker width = 1.5

    low_limit = Q1 - (whisker width * IQR)

    High_limit = Q3 + (whisker width * IQR)

```python
sns.boxplot(data=df, orient="h", palette="Set2")
```

```python
sns.boxplot(data = df , palette = 'BuPu' )
```

```python
violinplot : A combination of boxplot and kernel density estimation.
```

```python
sns.violinplot(data = df , palette = 'rocket')
```
