Q1. Explain GET and POST methods.

Ans:

HTTP stands for HyperText Transfer Protocol which has some set of rules which determines the communication 

or data transfer into client-server architecture. 

The client is generally a browser, and the Server is the source where information is available already, 

and the Client requests the Server for the information. A server is a computer where the website is hosted. 

A client or the browser sends a request to the server, and the server responds to the request of the client. 

There are a lot of methods used by HTTP for data transfer or establishing the communication between client and server, 

like GET, POST, PUT, HEAD, DELETE, PATCH, OPTIONS, TRACE, etc.

There are generally two methods used for the information transfer between client-server:

    GET method or request
    POST method or request


GET :

The GET method is used to request the server to GET the data or information.

It should not be used with the requesting sensitive data, and there are certain limits to the requesting data. 

GET requests to remain in the browser history, and it can also be bookmarked. GET requests can be cached.

To use the GET method, we should have installed the library which supports the GET request like urllib,httplib, or requests.

POST :

The POST method is used to send the data to the server. 

It is generally used when we want to upload data or any file to the server, like filling out a form, etc.

POST requests cannot be bookmarked, and it does not remain in the browser history. It cannot be cached.

It can be used with the sensitive and private data, and it has no limit for data length.
