#### Q1. Explain the concept of R-squared in linear regression models. How is it calculated, and what does it represent?

Ans:

#### R-squared : 
    
    1. R-squared is the accuracy of the model in linear regression.
    
    R-squared = 1 - (SSR/SST)
    
    SSR : sum of square of residual
        
    SST : sum of square of Total
        
        
    R-squared = 1 - (yact - ypred)^2 / (yact - yavg)^2
    
    yact : Actual output data points
        
    ypred : prediction output data points
        
    yavg : Average of Actual output data points
        
        
    R-squared value is between 0 to 1.
    
    R-squared value is always greater than Adjusted R^2.

#### Q2. Define adjusted R-squared and explain how it differs from the regular R-squared.

Ans:

#### Adjusted R-squared :
    
    1.Adjusted R-squared is the one of the cost function in linear regression.
    
    Adjusted R-squared =  1 - (1 - R-squared)*(N - 1) / (N - k - 1)
    
    R-squared = Accuracy of the model
    
    N = no. of records of dependent variable
    
    k = no. of independent variable
    
    Adjusted R squared is always less than R-squared

#### Q4. What are RMSE, MSE, and MAE in the context of regression analysis? How are these metrics calculated, and what do they represent?

Ans:

#### MSE : 
    
    1.Mean Squared Error is one of the performance matrix of the linear regression.
    
    mse = (yact - ypred)^2 / n
    
    yact = Actual output data points
    
    ypred = prediction output data points
    
    Mean squared error is  used when outliers are not present.
    

#### MAE : 
    
    1.Mean Absolute Error is one of the performance matrix of the linear regression.
    
    MAE = (yact - ypred) / n
    
    yact = Actual output data points
    
    ypred = prediction output data points
    
    Mean Actual Error is used when outliers are present.
    

#### RMSE :
    
    Root Mean Squared Error is one of the performance matrix of the linear regression.
    
    RMSE = np.sqrt(MSE)
    
    MSE = Mean Squared Error
